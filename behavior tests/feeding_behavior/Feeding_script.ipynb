Distance moved

```python
listdf1=[]
for df1 in listall:
    distance_moved=[] 
    for i in range(1,len(df1.X_center)):
        a = df1.X_center[i]- df1.X_center[i-1] #a = current x coordinate - previous x coordinate
        b = df1.Y_center[i]- df1.Y_center[i-1] #b = current y coordinate - previous y coordinate
        truedistance = sqrt(a**2 + b**2) # hypotenuse calculation
        distance_moved.append(truedistance)
    distance_moved.insert(0,0)
    df1['Distance_moved']=distance_moved # replace Distance_moved column with new true distance"
    listdf1.append(df1)
```

```python
### print out group data ###
listdepth,listdm,listfr,listtime,listedge,listglassedge,listglassedge1=[],[],[],[],[],[],[]

for df1 in listdf1:
    summ=df1.groupby('phase')['depth'].sum()
    length=df1.groupby('phase')['depth'].count()

    for i in range(1,11):
        depth=(summ[i]/length[i])
        listdepth.append(depth)
        listtime.append(i)

    summ=df1.groupby('phase')['Distance_moved'].sum()
    length=df1.groupby('phase')['Distance_moved'].count()
    for i in range(1,11):
        dm=(summ[i]/length[i])
        listdm.append(dm*30)
    
    summ=df1.groupby('phase')['freezing'].sum()
    length=df1.groupby('phase')['freezing'].count()
    for i in range(1,11):
        fr=(summ[i]/length[i])
        listfr.append(fr)

listofdates,listoftreats,listofstrains=[],[],[]
for df1 in listdf1:
    for i in range(1,11):
        listofdates.append(df1.date[0])
        listoftreats.append(df1.treatment[0])
        listofstrains.append(df1.strain[0])
        
tnt = pd.DataFrame({'time':listtime,'date':listofdates,'Treatment':listoftreats,'Strain':listofstrains,'depth':listdepth,'speed':listdm,'freezing':listfr})
tnt.to_csv('data/tnt.csv')
```
